```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("./2018CrimeStatisticsByCategoryOfCrime.csv")
df2 = pd.read_csv("./MPsProfileInformation10thParliament.csv", encoding="ISO-8859-1")
df3 = pd.read_csv("./UceResultsBySchool2011-2016.csv")
```

```python
df1.head()
```

```python
#Drop unnecessary row 0
df1 = df1.drop(0, axis = 0)
```

```python
df1.head()
```

```python
#The first two columns in the dataset have null values because there column labels are shifted below
#fix this by shifting these 2 column values back up

df1[["Unnamed: 0", "Unnamed: 1"]] = df1[["Unnamed: 0", "Unnamed: 1"]].shift(-1)
df1.head()
```

```python
#Now the actual data columns are aligned at row 1, 
#we have to create a new dataframe with the row 1 values as the column headings

df1_values = df1.iloc[2:]
new_Columns = df1.iloc[0].values
```

```python
new_Columns
```

```python
df1_values.head()
```

```python
#now create a new dataframe with those 2 values
new_df1 = pd.DataFrame(df1_values.values, columns=new_Columns)
new_df1.head()
```

```python
new_df1.fillna(0, inplace = True)
```

```python
new_df1.head()
```

## Correlation Matrix

```python
import re

def remove_invalid_chars(text):
    text = str(text)
    text = re.sub("[-,]","",text)
    return text

remove_invalid_chars("2,000")
```

```python
#remove all characters in integer columns so as to allow casting from object to integer
for column in new_df1:
    try:
        new_df1[column] = new_df1[column].map(remove_invalid_chars)

    except Exception:
        new_df1[column] = new_df1[column].applymap(remove_invalid_chars)
```

```python
#convert all the columns to numeric
for column in new_df1:
    try:
        new_df1[column] = pd.to_numeric(new_df1[column])

    except Exception as e:
        print(column)
        continue
```

```python
corr = new_df1.corr()

corr
```

```python
plt.matshow(corr)
```

```python
fig = plt.figure(figsize=(20,20))
ax = fig.gca()
new_df1.hist(ax = ax)
```

```python
# anajyszing df2

df2.head()
```

```python
df2.shape
```

```python
for column in df2:
    print(column, df2[column].isnull().sum())
```

```python
#dropping all columns with 200+ null values
for column in df2:
    count = df2[column].isnull().sum()

    if count > 200:
        df2 = df2.drop(column, axis = 1)
```

```python
df2.head(n=3)
```

```python
for column in df2:
    df2[column].fillna(df2[column].mode()[0], inplace = True) #replace null values with the most frequent values
```

```python
df2.head()
```

```python
#convert all the numerical columns to integer
for column in df2:
    try:
        df2[column] = pd.to_numeric(df2[column])
        print(column)

    except Exception:
        #can't be convered to numeric
        pass
```

```python
#only one numeric column
print(df2.columns)
```

```python
#encode the dataset into numerical values
columns_dict = dict()

for column in df2:
    columns_dict[column] = df2[column].unique()
```

```python
columns_dict
```

```python
#replace all ocuureneces of a value with its columns dict index
for column in df2:
    for value in columns_dict[column]:
        df2[column][df2[column] == value] = list(columns_dict[column]).index(value)
```

```python
df2.head()
```

```python
for column in df2:
    df2[column] = pd.to_numeric(df2[column])
```

```python
#finally the dataset is numeric 
corr = df2.corr()

corr
```

```python
import seaborn as sns

sns.heatmap(corr)
```

```python
fig = plt.figure(figsize=(20, 20))

ax = fig.gca()

df2.hist(ax = ax)
```
